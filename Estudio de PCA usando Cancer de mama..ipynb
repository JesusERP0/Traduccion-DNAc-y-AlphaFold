**Visualización de datos usando PCA**

PCA convierten datos de alta dimensión en un componente principal de baja dimensión. Empezamos por estandarizar los datos, normalizaremos usando StandardScaler.

```python
from sklearn.preprocessing import StandardScaler
x = mama_dataset.loc[:, Caracteristicas].values
x = StandardScaler().fit_transform(x) # Normalizando caract.
```

```python
x.shape
```

```python
#Comprobando si los datos normalizados tienen una media de cero (.mean) y una desviación estándar de uno(.std).
np.mean(x),np.std(x)
```

```python
#Convierto las funciones normalizadas a un formato tabular
Caract_cols = ['Caracteristica'+str(i) for i in range(x.shape[1])]
mama_normalizado = pd.DataFrame(x,columns=Caract_cols)
mama_normalizado.tail(10)
```

Las próximas líneas de código proyectarán los datos tridimensionales del cáncer de mama a componentes principales bidimensionales.

Utilizaremos la biblioteca sklearn para importar el módulo PCA y, en el método PCA, pasará la cantidad de componentes (n_components=2) y finalmente llamará a fit_transform en los datos agregados.


```python
from sklearn.decomposition import PCA
pca_mama = PCA(n_components=2)
principalComponents_mama = pca_mama.fit_transform(x)
```

```python
#creemos un DataFrame que tendrá los valores de los componentes principales para las 569 muestras.
principal_mama_Df = pd.DataFrame(data = principalComponents_mama
             , columns = ['Componente principal 1', 'Componente principal 2'])
```

```python
principal_mama_Df.tail(10)
```

```python
print('Variación explicada por componente principal: {}'.format(pca_mama.explained_variance_ratio_))
```

Del resultado anterior, se puede observar que el componente principal 1 contiene el 44.27% de la información, mientras que el componente principal 2 contiene solo el 18.97% de la información. Además, el otro punto a tener en cuenta es que al proyectar datos de treinta dimensiones a datos de dos dimensiones, se perdió el 36,8% de la información.

Tracemos la visualización de las 569 muestras a lo largo del eje componente principal 1, y componente principal 2. 

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Componente principal-1',fontsize=20)
plt.ylabel('Componente principal-2',fontsize=20)
plt.title("Análisis de componentes principales del conjunto de datos sobre cáncer de mama",fontsize=20)
targets = ['Benigno', 'Maligno']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = mama_dataset['Etiqueta'] == target
    plt.scatter(principal_mama_Df.loc[indicesToKeep, 'Componente principal 1']
               , principal_mama_Df.loc[indicesToKeep, 'Componente principal 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

En el gráfico anterior, se puede observar que las dos clases benignas y malignas, cuando se proyectan en un espacio bidimensional, pueden ser linealmente separables hasta cierto punto. Otras observaciones pueden ser que la clase benigna está más extendida en comparación con la clase maligna.
