```python
#Primer metodo con Biopython.
from Bio.Seq import Seq

# Función para traducir una secuencia de cDNA a aminoácidos
def traducir_cDNA(secuencia_cDNA):
    secuencia_cDNA = secuencia_cDNA.upper().replace('T', 'U')
    secuencia_cDNA = Seq(secuencia_cDNA)
    secuencia_aminoacidos = secuencia_cDNA.translate()
    return str(secuencia_aminoacidos)

# Ejemplo
secuencia_cDNA = 'ATGGTGCAGCTGAAATAA'
secuencia_aminoacidos = traducir_cDNA(secuencia_cDNA)
print("Secuencia de aminoácidos: ", secuencia_aminoacidos)
```

En este ejemplo, se utiliza la función translate() de Biopython para traducir la secuencia de cDNA directamente a la secuencia de aminoácidos. Antes de realizar la traducción, la secuencia de cDNA se convierte en un objeto de tipo Seq. Además, se reemplazan las bases de timidina (T) por uracilo (U) ya que la función translate() espera una secuencia de ARN mensajero (ARNm) como entrada.
